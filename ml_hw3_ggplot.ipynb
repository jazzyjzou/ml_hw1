```python
import numpy as np 
import pandas as pd 

from sklearn import preprocessing
import matplotlib.pyplot as plt 
plt.rc("font", size=14)
import seaborn as sns
sns.set(style="white") #white background style for seaborn plots
sns.set(style="whitegrid", color_codes=True)

fruits = pd.read_csv('fruit_color.csv')
fruits.head()
```

```python
sns.countplot(fruits['fruit_name'], label = 'count')
plt.show()
```

```python
from ggplot import *
ggplot(fruits, aes(x='width', y = 'height', colour ='mass')) +\
    geom_point() +\
    facet_wrap('fruit_subtype')
```

```python
pivot_fruits = pd.pivot_table(fruits, index='fruit_subtype', columns='fruit_name', values=['color_score'],aggfunc=np.sum)

sns.heatmap(pivot_fruits)
plt.xlabel('fruit_name')
plt.ylabel('fruit_subtype')
plt.title('name_subtype')
plt.show()
```
